# Momentum Gradient Descent Algorithm Implementation
![formaula](https://editor.analyticsvidhya.com/uploads/78368mom_eq.png)

```python
def momintam_GD_alg(X, Y, alpha = 0.01, gama=0.9, itr = 1e6, bias=True):
    # *************** Step 1 ***************
    if len(X.shape) > 1:
        m,n = X.shape
    else:
        m = X.shape[0]
        n = 1
    if bias:
        thetas = np.zeros((n+1, 1))  
        x0 = np.ones((m,1))
        X = np.c_[x0, X.values]
        Y = Y.values.reshape(-1,1)
    else:
        thetas = np.zeros((n, 1))  
        X = X.values
        Y = Y.values.reshape(-1,1)
    
    thetas_history = []
    loss_history = []
    m_t = 0
    
    for _ in range(int(itr)):
        # *************** Step 2 ***************
        y_hat = X @ thetas # X(n,m) @ thetas(n,1)


        # *************** Step 3 ***************
        error = y_hat - Y
        J = ((error.T @ error).item())/(2*m) # error.T(1,m) error(m,1) = (1,1)


        # ************ stop condtion ************
        if len(thetas_history) > 2:
            if np.abs(loss_history[-1]-loss_history[-2]) < 0.0001:
                break
        loss_history.append(J)
        thetas_history.append(thetas)

        
        # *************** Step 4 ***************
        G_thetas = (X.T @ error)/m  # X(n,m) @ error(m,1) = (n,1)

        m_t = gama * m_t + alpha * G_thetas
        # *************** Step 5 ***************
        thetas = thetas - m_t

    h = X @ thetas
    return h, thetas, thetas_history, loss_history
```

```python
h, ths, ths_hist, loss_hist = momintam_GD_alg(singel_data.iloc[:,0], singel_data.iloc[:,1], alpha = 0.01, itr = 1e6, bias=True)
```

```python
plt.figure(figsize=(10,5))
plt.plot(loss_hist)
plt.plot(loss_hist, 'o')
plt.figure(figsize=(10,5))
plt.plot(np.array(ths_hist)[:,0], loss_hist)
plt.plot(np.array(ths_hist)[:,0], loss_hist, 'o')
plt.figure(figsize=(10,5))
plt.plot(np.array(ths_hist)[:,1], loss_hist)
plt.plot(np.array(ths_hist)[:,1], loss_hist, 'o')

r2_score(singel_data.iloc[:,1], h)
```
